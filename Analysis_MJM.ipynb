```python
#This code was written for windows
#Import dependencies
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Upload csv files
MDA_file = "MDA_table/Master_Table_MDA.csv"
MTR_file = "MTR_table/Master_Table_MTR.csv"
```

```python
#Create Panda dataframes with workable names
df_MDA = pd.read_csv(MDA_file)
df_MDA = df_MDA.rename(columns={'PRECIO MDA': 'pz_MDA', 'PRECIO ENERGIA MDA': 'pz_ene_MDA', 
                                'PRECIO PERDIDA MDA': 'pzl_per_MDA', 'PRECIO CONGESTION MDA': 'pz_cng_MDA'})
df_MTR = pd.read_csv(MTR_file)
df_MTR = df_MTR.rename(columns={'PRECIO MTR': 'pz_MTR', 'PRECIO ENERGIA MTR': 'pz_ene_MTR', 
                                'PRECIO PERDIDA MTR': 'pzl_per_MTR', 'PRECIO CONGESTION MTR': 'pz_cng_MTR'})
```

```python
#Drop rows with NULL values
df_MDA = df_MDA.replace('NULL', np.nan).dropna()
df_MTR = df_MTR.replace('NULL', np.nan).dropna()
```

```python
#Start analysis for MDA prices
#Create pivot table to make a graph for average prices per day in all zones and systems

avg_daily_MDA = df_MDA.drop(["HORA"], axis=1)                                   
avg_daily_MDA = pd.pivot_table(avg_daily_MDA,index=["SISTEMA","ZONA","FECHA"],
               values=["pz_MDA","pz_ene_MDA","pzl_per_MDA", "pz_cng_MDA"],aggfunc=[np.mean])

avg_daily_MDA.head(400)

```

```python
#Filter DF by sistema
avg_daily_MDA_BCA = avg_daily_MDA.query('SISTEMA == ["BCA"]')
avg_daily_MDA_BCS = avg_daily_MDA.query('SISTEMA == ["BCS"]')
avg_daily_MDA_SIN = avg_daily_MDA.query('SISTEMA == ["SIN"]')


#Create empty vectors for each sistem (SIN, BCA, BCS) and price (pz, pz_ene, pzl_per, pz_cng)

x_SIN= []
y_SIN_pz= []
y_SIN_pz_ene= []
y_SIN_pzl_per= []
y_SIN_pz_cng= []

x_BCA = []
y_BCA_pz=[]
y_BCA_pz_ene= []
y_BCA_pzl_per= []
y_BCA_pz_cng= []

x_BCS=[]
y_BCS_pz= []
y_BCS_pz_ene= []
y_BCS_pzl_per= []
y_BCS_pz_cng= []

#Store data from the dataframe to the corresponding vector
#BCA
i = 1
    
for FECHA in avg_daily_MDA_BCA.to_records():
    y_BCA_pz.append(FECHA[3])
    y_BCA_pz_ene.append(FECHA[5])
    y_BCA_pzl_per.append(FECHA[6]) 
    y_BCA_pz_cng.append(FECHA[4])
    if i < 365:
        x_BCA.append(i)
        i=i+1
    else:
        x_BCA.append(i)
        i=1
#BCS
i = 1
    
for FECHA in avg_daily_MDA_BCS.to_records():
    y_BCS_pz.append(FECHA[3])
    y_BCS_pz_ene.append(FECHA[5])
    y_BCS_pzl_per.append(FECHA[6]) 
    y_BCS_pz_cng.append(FECHA[4])
    if i < 365:
        x_BCS.append(i)
        i=i+1
    else:
        x_BCS.append(i)
        i=1
#SIN
i = 1
    
for FECHA in avg_daily_MDA_SIN.to_records():
    y_SIN_pz.append(FECHA[3])
    y_SIN_pz_ene.append(FECHA[5])
    y_SIN_pzl_per.append(FECHA[6]) 
    y_SIN_pz_cng.append(FECHA[4])
    if i < 365:
        x_SIN.append(i)
        i=i+1
    else:
        x_SIN.append(i)
        i=1
```

```python
#Build the scatter plot for Zonal Prices (pz)
#plt.figure(figsize=(20,10))
plt.scatter(x_BCA,y_BCA_pz,facecolors="gold", edgecolors= "black", alpha=0.7, label = 'BCA')
plt.scatter(x_BCS,y_BCS_pz,facecolors="lightskyblue", edgecolors= "black", alpha=0.7, label = 'BCS')
plt.scatter(x_SIN,y_SIN_pz,facecolors="lightcoral", edgecolors= "black", alpha=0.7, label = "SIN")

# Incorporate the other graph properties
plt.title("Average Zonal Daily Prices (MDA, 2018)")
plt.xlabel("Day in the year")
plt.ylabel("PML per load zone [$MXN/MWh-day]")
plt.grid(True)

#Put limits
plt.ylim(-600, 9000)

#Create a legend
lgnd=plt.legend(title="Siystem")
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Save Figure
plt.savefig("Figures\Avg_Zonal_MDA.png")
```

```python
#Go deeper and analysis prices per hour
#Create pivot table to make a graph for average prices per hour in all zones and systems

avg_hour_MTR = df_MTR.drop(["FECHA"], axis=1)                                   
avg_hour_MTR = pd.pivot_table(avg_hour_MTR,index=["SISTEMA","ZONA","HORA"], values=["pz_cng_MTR","pz_MTR","pz_ene_MTR",
                                                                                    "pzl_per_MTR"])

avg_hour_MTR.head()

```

```python
#Filter DF by sistema
avg_hour_MTR_BCA = avg_hour_MTR.query('SISTEMA == ["BCA"]')
avg_hour_MTR_BCS = avg_hour_MTR.query('SISTEMA == ["BCS"]')
avg_hour_MTR_SIN = avg_hour_MTR.query('SISTEMA == ["SIN"]')

#Create empty vectors for each sistem (SIN, BCA, BCS) and price (pz, pz_ene, pzl_per, pz_cng)

x_SIN= []
y_SIN_pz= []
y_SIN_pz_ene= []
y_SIN_pzl_per= []
y_SIN_pz_cng= []

x_BCA = []
y_BCA_pz=[]
y_BCA_pz_ene= []
y_BCA_pzl_per= []
y_BCA_pz_cng= []

x_BCS=[]
y_BCS_pz= []
y_BCS_pz_ene= []
y_BCS_pzl_per= []
y_BCS_pz_cng= []

#Store data from the dataframe to the corresponding vector
#BCA
i = 1
    
for HORA in avg_hour_MTR_BCA.to_records():
    y_BCA_pz.append(HORA[3])
    y_BCA_pz_ene.append(HORA[5])
    y_BCA_pzl_per.append(HORA[6]) 
    y_BCA_pz_cng.append(HORA[4])
    if i < 24:
        x_BCA.append(i)
        i=i+1
    else:
        x_BCA.append(i)
        i=1
#BCS
i = 1
    
for HORA in avg_hour_MTR_BCS.to_records():
    y_BCS_pz.append(HORA[3])
    y_BCS_pz_ene.append(HORA[5])
    y_BCS_pzl_per.append(HORA[6]) 
    y_BCS_pz_cng.append(HORA[4])
    if i < 24:
        x_BCS.append(i)
        i=i+1
    else:
        x_BCS.append(i)
        i=1
#SIN
i = 1
    
for HORA in avg_hour_MTR_SIN.to_records():
    y_SIN_pz.append(HORA[3])
    y_SIN_pz_ene.append(HORA[5])
    y_SIN_pzl_per.append(HORA[6]) 
    y_SIN_pz_cng.append(HORA[4])
    if i < 24:
        x_SIN.append(i)
        i=i+1
    else:
        x_SIN.append(i)
        i=1
```
